### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array = np.random.randint(1, 245, size=(7,7))

avr = array.mean()
std_otkl = np.nanstd(array)

#Нормализация
norm_array = np.linalg.norm(array)
unit_array = array / norm_array

print(f'среднее значение: {avr}')
print(f'стандартное отклонение: {std_otkl}')
print(f'длина нормальзованного вектора: {norm_array}')
print(f'единичный нормализованный вектор: {unit_array}')
print(f'нормализованный вектор: {norm_array} х {unit_array}')
```

# Работа над ошибками

В идеале, любой массив должен приводиться примерно к одному масштабу - это очень полезно для обучения линейных моделей, который признакам присваивают веса (если все признаки одного масштаба, веса будут давать информацию о вкладе признаков в прогноз). 

!! Заданием подразумевается вариант вычесть из матрицы среднее и поделить на стандартное отклонение (можно также проверить на тех же матрицах, какие получатся элементы, какой у них будет масштаб)

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array = np.random.randint(1, 245, size=(7,7))

avr = array.mean()
std_otkl = np.nanstd(array)

#Нормализация
norm_array = np.linalg.norm(array)
unit_array = (array - avr)/std_otkl

print(f'среднее значение: {avr}')
print(f'стандартное отклонение: {std_otkl}')
print(f'длина нормальзованного вектора: {norm_array}')
print(f'нормализованный вектор: {unit_array}')
print(f'вектор: {array}')
```
